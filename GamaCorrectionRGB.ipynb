```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt
from PIL import Image
```

```python
image = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Sunrise.jpg')
```

```python
resize_image=cv2.resize(image,(600,250))
```

```python
brightness_factor = 1.2 #load a blurred noisy image
adjusted_image = cv2.convertScaleAbs(resize_image,alpha=brightness_factor)
cv2_imshow(resize_image)
cv2_imshow(adjusted_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
blue_channel, green_channel, red_channel = cv2.split(adjusted_image)
```

```python
gamma_blue = 1.3
gamma_green = 1.3
gamma_red = 1.1
```

```python
blue_corrected = np.power(blue_channel / 255.0, gamma_blue) * 255.0
green_corrected = np.power(green_channel / 255.0, gamma_green) * 255.0
red_corrected = np.power(red_channel / 255.0, gamma_red) * 255.0
```

```python
blue_corrected = blue_corrected.astype(np.uint8)
green_corrected = green_corrected.astype(np.uint8)
red_corrected = red_corrected.astype(np.uint8)
```

```python
gamma_corrected_image = cv2.merge((blue_corrected, green_corrected, red_corrected))
```

```python
cv2.imwrite('gamma_corrected_rgb_image.jpg', gamma_corrected_image)  # Replace with the desired file path
```

```python
print('original')
cv2_imshow(resize_image)
print('blur noisy')
cv2_imshow(adjusted_image)
print('gama correction')
cv2_imshow(gamma_corrected_image )
cv2.waitKey(0)
cv2.destroyAllWindows()
```
